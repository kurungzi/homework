```python
import random

class Hangman:
    def __init__(self, words):
        self.words = words
        self.secret_word = ""
        self.guesses = []
        self.max_attempts = 6
        self.remaining_attempts = self.max_attempts

    def choose_word(self):
        self.secret_word = random.choice(self.words).lower()

    def display_word(self):
        displayed_word = ""
        for letter in self.secret_word:
            if letter in self.guesses:
                displayed_word += letter + " "
            else:
                displayed_word += "_ "
        return displayed_word.strip()

    def guess_letter(self, letter):
        letter = letter.lower()
        if letter in self.guesses:
            print("이미 단어를 맞추셨습니다", letter)
            return
        
        self.guesses.append(letter)
        if letter not in self.secret_word:
            self.remaining_attempts -= 1

    def get_status(self):
        if self.remaining_attempts <= 0:
            return "실패"
        elif all(letter in self.guesses for letter in self.secret_word):
            return "성공"
        else:
            return "ongoing"

    def print_game_status(self):
        print("현재 단어:", self.display_word())
        print("Guesses:", ", ".join(self.guesses))
        print("남은 기회:", self.remaining_attempts)

# 단어 목록을 설정
word_list = ["apple", "banana", "cherry", "date", "grape"]

# Hangman 객체를 생성하고 게임을 실행
hangman_game = Hangman(word_list)
hangman_game.choose_word()

print("게임을 시작합니다!")
print("단어를 맞춰보세요", hangman_game.max_attempts, "기회")

while hangman_game.get_status() == "진행 시작":
    hangman_game.print_game_status()
    guess = input("단어를 추측해보세요 : ")
    hangman_game.guess_letter(guess)

if hangman_game.get_status() == "승리":
    print("축하합니다! 단어를 맞추셨습니다.", hangman_game.secret_word)
else:
    print("안 돼, 돌아가", hangman_game.secret_word)
```
